```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
battle = pd.read_csv('battles.csv')
battle.head()
```

```python
battle.shape
```

```python
battle.rename(columns={'attacker_1':'primary_attacker'},inplace=True)
battle.head()
```

```python
battle.rename(columns={'defender_1':'primary_defender'},inplace=True)
battle.head()
```

```python
battle['attacker_king'].value_counts()   #no. of times a king has attacked
```

```python
battle['location'].value_counts()   #no. of attacks in locations
```

```python
sns.set(rc={'figure.figsize':(13,5)})   #bargraph using seaborn for a king and his army
sns.barplot(x='attacker_king',y='attacker_size',data=battle)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(13,5)})    #bargraph for a defender and his army
sns.barplot(x='defender_king',y='defender_size',data=battle)
plt.show()
```

```python
sns.countplot(x=battle['attacker_king'],hue=battle['battle_type'])
plt.show()
```
